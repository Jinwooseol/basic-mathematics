특정 구간 내에서 min value 구하기

```python
def get_min(x1, x2, f):
  x = np.arange(x1, x2, 0.01)
  y = f(x)

  plt.plot(x, y)
  plt.xlabel('x')
  plt.ylabel('y')
  plt.title('get minimum')
  plt.show()

  return min(y)
```

```python
print(get_min(1, 4, my_func))
```

Differentiation
- $\frac{df(x)}{dx} = lim_{x->\infty}\frac{f(x+h) - f(x-h)}{2h}$
- h는 아주 작은 수를 의미하나 너무 작은 수의 경우 python은 0.0으로 인식하기 때문에 딥러닝에서는 1$e$ - 4정도로 설정

```python
def numerical_differential(f, x):
  h = 1e-4
  return (f(x+h)-f(x-h)) / (2*h)
```

Gradient
- 방향성을 가짐

```python
def numerical_diff(f, x):
  h = 1e-5
  grad = np.zeros_like(x) # x와 모양은 같으나 0으로 채워진 array

  for i in range(x.size):
    tmp = x[i]

    x[i] = tmp + h
    fxh1 = f(x)

    x[i] = tmp - h
    fxh2 = f(x)

    grad[i] = (fxh1 - fxh2) / (2*h)
    x[i] = tmp
  
  return grad
```

```python
print(numerical_diff(partial_derivative, np.array([3.0, 4.0])))
print(numerical_diff(partial_derivative, np.array([1.0, 2.0])))
```
